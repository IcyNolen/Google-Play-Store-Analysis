<span style="font-size: 24px;">Basics</span>

```python
import pandas as pd
data = pd.read_csv("googleplaystore.csv")
data
```

```python
print(data.isnull().sum())
#returns the no of null values in each columns.
```

```python
df=data.dropna()
#deletes the specific rows containing nan values
print(df.isnull().sum())
```

```python
df.head()
#return top 5 rows.
```

<span style="font-size: 24px;">Finding average rating</span>

```python
avg= int( sum(df['Rating']) / len(df['Rating']) )
print(avg)      
```

<span style="font-size: 24px;">Apps with Rating between the range of 4 and 4.5</span>


```python
c=0
for i in df['Rating']:
    if (i>=4.0 and i<=4.5):
        c+=1
print(c)
```

<span style="font-size: 24px;">Total number of apps in each category</span>


```python
categories = {}

for name in df['Category'].unique():
    ct = 0
    for i in df['Category']:
        if(i == name):
            ct += 1
    categories[name] = ct
    
print(categories)
```

<span style="font-size: 24px;">To replace all null values with most frequent values</span>

```python
import numpy as np
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df.iloc[:,:]=imputer.fit_transform(df.iloc[:,:].values)
df
```

```python
df.describe()
```

<strong style="font-size: 24px;">How many free apps are there in ART_AND_DESIGN?</strong>>


```python
df1 = df[df['Category'] == 'ART_AND_DESIGN']
print(len(df1[df1['Type'] == 'Free']))
```

<strong style="font-size: 24px;"> Which category is having maximum average rating Descending order?</strong>


```python
df_copy = df.copy()
# Convert the 'Rating' column to numeric without modifying the original DataFrame
df_copy['Rating'] = pd.to_numeric(df_copy['Rating'])

# Now group by 'Category', calculate the mean rating, and sort in descending order
result = df_copy.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print(result)

```
